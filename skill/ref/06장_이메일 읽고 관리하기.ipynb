```python
import imapclient
import pyzmail

imap = imapclient.IMAPClient('imap.gmail.com', ssl=True)
imap.login('my_email', 'my_passcode')  # id, password
imap.select_folder('INBOX', readonly=False)
uids = imap.gmail_search('has:attachment')[:3]  # 첨부파일 있는 메일 3 개만 검색

for uid in uids:
    rmsgs = imap.fetch([uid], ['BODY[]'])
    message = pyzmail.PyzMessage.factory(rmsgs[uid][b'BODY[]'])
    subject = message.get_subject()
    senders = message.get_addresses('from')

    print('제목:', message.get_subject())    # 제목 출력
    # 첨부 파일 저장
    for part in message.mailparts:
        if part.filename:
            print('saving..', part.filename, part.type)
            cont = part.get_payload()
            if part.type.startswith('text/'):
                open(part.filename, 'w').write(cont)
            else:
                open(part.filename, 'wb').write(cont)
    print()
    
imap.logout();
```

```python
import re

fnames = []

for uid in uids:
    rmsgs = imap.fetch([uid], ['BODY[]'])
    m = rmsgs[uid][b'BODY[]']
    message = pyzmail.PyzMessage.factory(m)
    subject = message.get_subject()
    senders = message.get_addresses('from')

#     print('제목:', message.get_subject())    # 제목 출력
    # 첨부 파일 저장
    for part in message.mailparts:
        if part.filename and part.filename.endswith('.pdf'):
            cont = part.get_payload()
            year, month = re.search('(\d{4})(\d{2})', part.filename).groups()
            filename = f'{year}_{month}_요금명세서.pdf'
            print('[saving]', filename)
            open(filename, 'wb').write(cont)
            fnames.append(filename)
    
imap.logout();
```

```python
import pikepdf
from tika import parser

for filename in fnames:
    year, month = re.search('(\d{4})_(\d{2})', filename).groups()
    pdf = pikepdf.open(filename, password='640115')  # 암호화된 파일을 읽어서
    pdf.save('output.pdf')                           # 암호가 풀린 상태로 저장한다
    raw = parser.from_file('output.pdf')   # tika 모듈로 pdf 파일을 분석한다
    content = raw['content']               # 본문 text 내용을 읽어낸다
    m = re.search('총(.*)?원 입니다', content)  # 필요한 총액만 추출한다
    charge = int(m.groups()[0].replace(',', ''))
    print('{}년 {}월 요금: {}'.format(year, month, charge))
```
