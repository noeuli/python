```python
import os

def get_list(fpath):
    for fname in os.listdir(fpath):
        print(fname)

get_list('C:\\Windows\\System32\\Drivers\\etc')
```

```python
for (curdir, dirs, files) in os.walk('.'): # (방문 디렉터리 경로, 디렉터리 목록, 파일 목록)
    print('curdir=', curdir) # 현재 방문중인 디렉터리
    print('dirs=', dirs[:5]) #디렉터리 이름들, 5개까지
    print('files=', files)
    print('-'*60)
```
